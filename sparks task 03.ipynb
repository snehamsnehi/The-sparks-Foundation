#### Removing irrelevant column

```python
col=['Postal Code']
s_store1=s_store.drop(columns=col,axis=1)
```

## Visualization

```python
print(s_store1['State'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=s_store1['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
print(s_store['Sub-Category'].value_counts())
plt.figure(figsize=(12,6))
sns.countplot(x=s_store['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("Region")
s_store['Region'].value_counts().plot.pie(autopct ='% 1.1f %%', shadow = True);
```

```python
plt.figure(figsize=(15,8))
plt.title("Segment")
s_store['Segment'].value_counts().plot.pie(autopct ='% 1.1f %%', shadow = True);
```

```python
plt.figure(figsize=(15,8))
plt.title("Category")
s_store['Category'].value_counts().plot.pie(autopct ='% 1.1f %%', shadow = True);
```

```python
n_data=['Sales','Quantity','Discount','Profit']
plt.figure(figsize=(10,5))
sns.heatmap(s_store[n_data].corr(),annot=True, fmt='.4f',cmap='RdYlGn',center=0 );
```

```python
figsize=(15,10)
sns.pairplot(s_store1,hue='Sub-Category')
```
