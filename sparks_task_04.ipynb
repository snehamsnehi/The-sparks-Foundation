## Data Visualization

```python
plt.figure(figsize=(15,8))
sns.countplot(x=df['iyear'], hue='success', data=df)
plt.ylabel('Number of attacks')
plt.xticks(rotation=90)
plt.xlabel('Attack Year')
plt.title('Increase in Terror',size=25, fontweight="bold")
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=df['country_txt'].value_counts()[:20].index,y=df['country_txt'].value_counts()[:20].values)
plt.ylabel('Number of attacks')
plt.xticks(rotation=90)
plt.xlabel('Attack Region')
plt.title('Country wise Attack',size=25, fontweight="bold")
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x=df['region_txt'], hue='success', data=df)
plt.ylabel('Number of attacks')
plt.xticks(rotation=90)
plt.xlabel('Attack Region')
```

```python
sns.barplot(x=df['attacktype1_txt'].value_counts()[:5].index,y=df['attacktype1_txt'].value_counts()[:5].values)
plt.ylabel('Total Attack', fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('Attack Type', fontsize=20)
plt.title('Attack types',size=25, fontweight="bold")
```

```python
sns.barplot(x=df['targtype1_txt'].value_counts()[:5].index,y=df['targtype1_txt'].value_counts()[:5].values)
plt.ylabel('Total Target', fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('Target Type', fontsize=20)
plt.title('Target types',size=25, fontweight="bold")
```

```python
sns.barplot(x=df['weaptype1_txt'].value_counts()[:5].index,y=df['weaptype1_txt'].value_counts()[:5].values)
plt.ylabel('Total weapon', fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('Weapon Type', fontsize=20)
plt.title('Weapon types',size=25, fontweight="bold")
```
